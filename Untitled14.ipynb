```python
def pojavljivanja(lista):
    pojavljivanja_autora = {}
    for i in lista:
        if i not in pojavljivanja_autora:
            pojavljivanja_autora[i] = 1
        else:
            pojavljivanja_autora[i] = pojavljivanja_autora[i]+1
    return_list = {i: v for i,v in sorted(pojavljivanja_autora.items(),key= lambda item: item[1], reverse = True)}
    return list(return_list.keys())[0:2]
    
izdavaci2= df.groupby('publisher')
publicationwise_top_authors = dict()

for a, grupisano in izdavaci2:
    svi_autori=[author for grupa in grupisano['authors'].values for author in grupa]
    najbolji_autori = pojavljivanja(svi_autori)
    publicationwise_top_authors[a] = najbolji_autori

najduze=max(len(v) for v in publicationwise_top_authors.values())

for key in publicationwise_top_authors:
    publicationwise_top_authors[key] += [None]*(najduze - len(publicationwise_top_authors[key]))
    
df2= pd.DataFrame(publicationwise_top_authors).T
df2.columns = ['Autor1','Autor 2']
df2.head()
```

```python
nltk.download('stopwords')
nltk.download('punkt')
```

```python
def proces_title(title):
    words=word_tokenize(title)
    words = [word.lower() for word in words]
    stop_words = set(stopwords.words('english'))
    words = [word for word in words if word not in stop_words and len(word) >1]
    return words

df['title_words']=df['title'].apply(proces_title)
all_words= [word for title_words in df['title_words'] for word in title_words]
word_counts = Counter(all_words)

najcesce = word_counts.most_common(10)
print('Najcesce kljucne reci su:\n')
print(f"{'Rec':<15}|{'Pojavljivanje':<10}")
print('-'*30)
for rec, n in najcesce:
    print(f"{rec:<15}|{n:<10}")
```
